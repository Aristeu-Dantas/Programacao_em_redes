## O problema do Troco

Faça um  programa que informa o número de cédulas de cada valor de real a serem retornadas a um cliente de um restaurante que pagou uma conta. O programa recebe o valor da conta e o valor que o cliente pagou; exibe o número de cédulas de cada valor que deve ser dado ao cliente. Suponha que o caixa tem número suficiente de cédulas de todos os valores.


## O problema do Troco - v2

Mas agora temos outro problema. O caixa não tem cédulas de
ilimitadas de todos os valores. Assim, antes de iniciar o processamento do
troco, solicite a quantidade de cédulas disponível de cada valor.
Só informe o troco se houver cédulas suficientes; caso contrário,
apresente a mensagem: 

**```O caixa não tem cédulas disponíveis para o troco!!!```**


````
def iniciaCaixa():
    global caixa
    print ("Inicialização do caixa")
    
def caixaTemCedulas(valor):
    asCedulas = []    

    if valor > 0: 
        return None
    else:
        return asCedulas
    
cedulas = (200, 100, 50, 20, 10, 5, 2,  1)
caixa   = [ 0 ] * len(cedulas)
iniciaCaixa()
print (cedulas)
print (caixa)

valorConta = int (input('Qual o valor da conta R$:'))
valorPago = int (input('Qual o valor pago R$:'))
troco = valorPago - valorConta
if troco > 0:
    print (f"Valor do troco: R$ {troco}")
    cedulasTroco = caixaTemCedulas(troco)
    if cedulasTroco != None:
        print ("Cédulas de troco:")
        for cedula in range(len(cedulas)):
            if cedulasTroco[cedula] > 0:
                print (f"# cedulas {cedulas[cedula]}: {cedulasTroco[cedula]}")
    else:
        print ("O caixa não tem cédulas para o troco...") 
elif troco == 0:
    print ("Pagamento sem troco...")
else:
    print ("Pagamento insuficiente...")
````

## O problema do Troco - v3
Além de verificar a a quantidade de cédulas disponíveis de cada valor, ser forem suficientes, diminua-as do caixa, indicando que o troco foi dado. Ao final o caixa terá o que tinha originalmente diminuído das cédulas de troco.

##O problema do Troco - v4

Faça com que seu programa **tente outras formas de devolver o troco**, quando certa combinação de cédulas não estiver disponível no caixa. 

Considere a seguinte situação do caixa:
```
----------------------------------------
Situação do caixa - Cédulas:
----------------------------------------
# de cédulas de R$ 100: 2
# de cédulas de R$  50: 1
# de cédulas de R$  20: 3
# de cédulas de R$   5: 5
# de cédulas de R$   2: 6
# de cédulas de R$   1: 2
```

Para um troco de  R$ 218, a proposição inicial seria de:

```
Troco - Cédulas:
----------------------------------------
# de cédulas de R$ 100:  2
# de cédulas de R$  10:  1
# de cédulas de R$   5:  1
# de cédulas de R$   2:  1
# de cédulas de R$   1:  1
----------------------------------------
```
Ocorre que não há notas de R$ 10 no caixa, gerando a mensagem:

**```O caixa não tem cédulas disponíveis para o troco!!!```**

Entretanto, é possível dar o troco, desde que as cédulas sejam:
```
----------------------------------------
# de cédulas de R$ 100:  1
# de cédulas de R$  50:  1
# de cédulas de R$  20:  3
# de cédulas de R$   2:  4
----------------------------------------
```

Desenvolva o programa considerando essa nova possibilidade. Ou seja, buscando em todas as alternativas de troco, uma possível.
